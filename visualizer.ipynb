```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from tensorflow.keras.datasets import cifar10

# Load CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
y_train = y_train.flatten()
y_test = y_test.flatten()

# Class names in CIFAR-10
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

# Plot sample images for each class
def plot_sample_images(x, y, class_names):
    plt.figure(figsize=(10, 5))
    for i, class_id in enumerate(np.unique(y)):
        idx = np.where(y == class_id)[0][0]
        plt.subplot(2, 5, i + 1)
        plt.imshow(x[idx])
        plt.title(class_names[class_id])
        plt.axis('off')
    plt.suptitle("Sample Images from CIFAR-10 Classes", fontsize=16)
    plt.tight_layout()
    plt.show()

# Plot class distribution histogram
def plot_class_distribution(y, class_names, dataset_name='Train'):
    sns.set(style='whitegrid')
    plt.figure(figsize=(10, 5))
    sns.histplot(y, bins=np.arange(11)-0.5, discrete=True, shrink=0.8)
    plt.xticks(ticks=np.arange(10), labels=class_names, rotation=45)
    plt.title(f'{dataset_name} Set Class Distribution')
    plt.xlabel('Class')
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()

# Run visualizations
plot_sample_images(x_train, y_train, class_names)
plot_class_distribution(y_train, class_names, 'Train')
plot_class_distribution(y_test, class_names, 'Test')
```
